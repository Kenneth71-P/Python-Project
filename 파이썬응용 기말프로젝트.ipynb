```python
#천안 2019~2021년도 최고 기온 분석

import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
font_path = "C:\Windows\Fonts\gulim.ttc"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)

nyc_temp_2019 = [4.5, 6.5, 13.3, 17.5, 25, 26.4, 28.6, 30.3, 25.4, 20.2, 13.8, 6.2]   #2019년의 천안시의 최고기온 리스트

nyc_temp_2020 = [6.7, 7.9, 13.9, 17.3, 23.5, 28.1, 26.6, 30.1, 24.9, 19.2, 13, 4.7]   #2020년의 천안시의 최고기온 리스트

nyc_temp_2021 = [3.3, 8.7, 14.9, 19.5, 21.7, 27.3, 31.2, 28.8, 25.9, 20.5, 13.8, 6.1] #2021년의 천안시의 최고기온 리스트

months = range(1,13)  #1월부터 12월까지
plt.title('천안 2019~2021년도 최고 기온 분석')
plt.xlabel('월')   # X축:월
plt.ylabel('온도') # Y축:온도
plt.plot(months, nyc_temp_2019,label='2019')  # 파란색 그래프
plt.plot(months, nyc_temp_2020,label='2020')  # 노란색 그래프
plt.plot(months, nyc_temp_2021,label='2021')  # 초록색 그래프

plt.legend()
```
